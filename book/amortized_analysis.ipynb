```python
class DynamicTable:
    def __init__(self):
        self.table = []
        self.size = 0   # capacity
        self.num = 0    # number of elements

    def table_insert(self, x):
        if self.size == 0:
            self.table = [None] * 1
            self.size = 1

        if self.num == self.size:
            new_table = [None] * (2 * self.size)

            # copy elements
            for i in range(self.num):
                new_table[i] = self.table[i]

            self.table = new_table
            self.size *= 2

        self.table[self.num] = x
        self.num += 1

    def __str__(self):
        return str(self.table[:self.num])
```

```python
T = DynamicTable()

for i in range(10):
    T.table_insert(i)
    print("Inserted:", i)
    print("Elements:", T)
    print("Capacity:", T.size)
    print()
```
