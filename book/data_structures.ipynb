```python
def stack_empty(s):
    return s.top == -1

def push(s, x):
    if s.top == len(s.data) - 1:
        raise Exception("overflow")
    s.top += 1
    s.data[s.top] = x

def pop(s):
    if stack_empty(s):
        raise Exception("underflow")
    x = s.data[s.top]
    s.top -= 1
    return x
```

```python
def enqueue(q, x):
    if (q.tail + 1) % len(q.data) == q.head:
        raise Exception("overflow")
    q.data[q.tail] = x
    q.tail = (q.tail + 1) % len(q.data)

def dequeue(q):
    if q.head == q.tail:
        raise Exception("underflow")
    x = q.data[q.head]
    q.head = (q.head + 1) % len(q.data)
    return x
```

```python
def list_search(l, k):
    x = l.head
    while x != None and x.key != k:
        x = x.next
    return x

def list_prepend(l, x):
    x.next = l.head
    x.prev = None
    if l.head != None:
        l.head.prev = x
    l.head = x

def list_insert(x, y):
    x.next = y.next
    x.prev = y
    if y.next != None:
        y.next.prev = x
    y.next = x

def list_delete(l, x):
    if x.prev != None:
        x.prev.next = x.next
    else:
        l.head = x.next
    if x.next != None:
        x.next.prev = x.prev
```

```python
"""
circular doubly linked
with sentinel

(use sentinels sparyling)
simplify code, but can waste memory
"""
def list_delete_double(x):
    x.prev.next = x.next
    x.next.prev = x.prev

def list_insert_double(x, y):
    x.next = y.next
    x.prev = y
    y.next.prev = x
    y.next = x

def list_search_double(l, k):
    l.nil.key = k
    x = l.nil.next
    while x.key != k:
        x = x.next
    if x == l.nil:
        return None
    else:
        return x
```

```python
def reverse_list(head):
    prev = None
    curr = head

    while curr is not None:
        next_node = curr.next
        curr.next = prev
        prev = curr
        curr = next_node

    return prev # new head
```

```python
"""
XOR properties:

a XOR a = 0
a XOR 0 = a
(a XOR b) XOR a = b

At node curr, if you know prev, you can compute next:
next = curr.np XOR prev

Reversing the list in O(1) time (swap head and tail)
In an XOR linked list:
Each node stores prev XOR next
Reversing the list just swaps the interpretation of prev and next
"""
def search(head, k):
    prev = 0
    curr = head

    while curr != 0:
        if curr.key == k:
            return curr
        next = curr.np ^ prev
        prev = curr
        curr = next

    return None

def insert_head(head, x):
    x.np = head  # since NIL = 0

    if head != 0:
        head.np = head.np ^ x

    return x  # new head
    
def delete(prev, x):
    next = x.np ^ prev

    if prev != 0:
        prev.np = prev.np ^ x ^ next

    if next != 0:
        next.np = next.np ^ x ^ prev
```

```python
"""
Provides a little better search than just using linear because
we provide a ranodom jump on a linked list

randomization can beat deterministic bounds
improve linked-list search without extra structure
(sorted increasing order)

O(sqrt(n))
"""
import random

def compact_list_search(key, next_idx, head, n, k):
    i = head
    
    while i is not None and key[i] < k:
        j = random.randint(0, n-1)
        
        if key[i] < key[j] and key[j] <= k:
            i = j
            
        if key[i] == k:
            return i
        
        i = next_idx[i]
        
    if i is not None and key[i] == k:
        return i

    return None

"""
Separates the randomness
instead of interleaves it as above

Guaranteed progress Only in phase 2
Less stable and depends on t
"""
def compact_list_search_bounded(key, next_idx, head, n, k, t):
    i = head

    for _ in range(0, t):
        j = random.randint(0, n-1)

        if key[i] < key[j] and key[j] <= k:
            i = j

            if key[i] == k:
                return i

    while i is not None and key[i] < k:
        i = next_idx[i]

    if i == None or key[i] > k:
        return None
    else:
        return i
```

```python
def linear_probing_hash_delete(t, q, hash_func):
    m = len(t)

    while True:
        t[q] = None
        q_prime = q

        while True:
            q_prime = (q_prime + 1) % m
            k_prime = t[q_prime]

            if k_prime is None:
                return

            home = hash_func(k_prime)

            if (home <= q < q_prime) or \
               (q_prime < home <= q) or \
               (q < q_prime < home):
                break

        t[q] = k_prime
        q = q_prime
```

```python
def swap(x, w):
    half = w // 2
    mask = (1 << w) - 1
    return ((x >> half) | (x << half)) & mask

def f_a(k, a, w):
    mod = 1 << w
    return swap((2 * k * k + a * k) % mod, w)

def f_a_pow(k, a, r, w):
    for _ in range(r):
        k = f_a(k, a, w)
    return k
```

```python
import math

def chop(k, w, u):
    mask = (1 << w) - 1
    return [(k >> (i * w)) & mask for i in range(u)]

def wee_hash(k, a, b, t, r, m, w=64):
    """
    WEE hash function

    k : integer key
    a : random odd integer in [0, 2^w)
    b : random seed in [0, 2^w)
    t : number of key bits
    r : number of iterations of f per word
    m : hash table size
    w : word size (default 64)
    """

    u = math.ceil(t / w)
    k_words = chop(k, w, u)

    q = b
    for i in range(u):
        q = f_a_pow(k_words[i] + q, a, r, w)

    return q % m
```
