```python
import random

def random_search(a, x):
    n = len(a)
    checked = set()

    while len(checked) < n:
        i = random.randrange(n)
        checked.add(i)

        if a[i] == x:
            return i

    return None

print(random_search([1, 4, 2, 0, 42], 3))
print(random_search([21, 553, 3, 36, 432, 1009], 3))
```
