```python
"""
Adjacency list → work proportional to actual edges
Adjacency matrix → work proportional to all possible edges

Original:
1: 2, 3
2: 3
3: 

Transpose:
1:
2: 1
3: 1, 2
"""
def transpose_adj_list(adj):
    V = len(adj)
    
    adjT = [[] for _ in range(V)]
    
    for u in range(V):
        for v in adj[u]:
            adjT[v].append(u)  # reverse edge
    
    return adjT

def transpose_matrix(A):
    V = len(A)
    
    AT = [[0] * V for _ in range(V)]
    
    for i in range(V):
        for j in range(V):
            AT[j][i] = A[i][j]
    
    return AT

def transpose_in_place(A):
    V = len(A)
    for i in range(V):
        for j in range(i + 1, V):
            A[i][j], A[j][i] = A[j][i], A[i][j]
```

```python
"""
input
0: 1, 1, 2, 0
1: 0, 2
2: 1
duplicate edges: (0,1), (0,1)
self-loop: (0,0)

output
0: 1, 2
1: 0, 2
2: 0, 1
"""
def simplify_to_undirected(adj):
    V = len(adj)
    adj_prime = [[] for _ in range(V)]
    
    seen = set()
    
    for u in range(V):
        for v in adj[u]:
            
            if u == v:
                continue # self-loops
            
            edge = (min(u, v), max(u, v))
            
            if edge not in seen:
                seen.add(edge)
                adj_prime[u].append(v)
                adj_prime[v].append(u)
    
    return adj_prime
```

```python
def square_adj_list(adj):
    V = len(adj)
    adj2 = [set() for _ in range(V)]
    
    for u in range(V):
        for v in adj[u]:
            adj2[u].add(v)
            
            for w in adj[v]:
                adj2[u].add(w)
    
    return [list(neighbors) for neighbors in adj2]
```

```python
def square_matrix(A):
    V = len(A)
    A2 = [[0]*V for _ in range(V)]
    
    for i in range(V):
        for j in range(V):
            if A[i][j] == 1:
                A2[i][j] = 1  # distance 1
            
            else:
                for k in range(V):
                    if A[i][k] == 1 and A[k][j] == 1:
                        A2[i][j] = 1
                        break
    
    return A2
```

```python
"""

In a directed graph, a universal sink is a vertex that:
has no outgoing edges (points to nobody)
has incoming edges from everyone else

In matrix terms (for vertex s):
Row s = all 0s
Column s = all 1s (except diagonal)
"""
def find_candidate(A):
    n = len(A)
    i, j = 0, 0
    
    while i < n and j < n:
        if A[i][j] == 1:
            i += 1
        else:
            j += 1
    
    return i if i < n else -1

def is_universal_sink(A, c):
    n = len(A)
    
    for j in range(n):
        if A[c][j] != 0:
            return False
    
    for i in range(n):
        if i != c and A[i][c] != 1:
            return False
    
    return True

def universal_sink(A):
    c = find_candidate(A)
    
    if c == -1:
        return None
    
    return c if is_universal_sink(A, c) else None
```
