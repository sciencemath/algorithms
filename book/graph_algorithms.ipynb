```python
"""
Adjacency list → work proportional to actual edges
Adjacency matrix → work proportional to all possible edges

Original:
1: 2, 3
2: 3
3: 

Transpose:
1:
2: 1
3: 1, 2
"""
def transpose_adj_list(adj):
    V = len(adj)
    
    adjT = [[] for _ in range(V)]
    
    for u in range(V):
        for v in adj[u]:
            adjT[v].append(u)  # reverse edge
    
    return adjT

def transpose_matrix(A):
    V = len(A)
    
    AT = [[0] * V for _ in range(V)]
    
    for i in range(V):
        for j in range(V):
            AT[j][i] = A[i][j]
    
    return AT

def transpose_in_place(A):
    V = len(A)
    for i in range(V):
        for j in range(i + 1, V):
            A[i][j], A[j][i] = A[j][i], A[i][j]
```

```python
"""
input
0: 1, 1, 2, 0
1: 0, 2
2: 1
duplicate edges: (0,1), (0,1)
self-loop: (0,0)

output
0: 1, 2
1: 0, 2
2: 0, 1
"""
def simplify_to_undirected(adj):
    V = len(adj)
    adj_prime = [[] for _ in range(V)]
    
    seen = set()
    
    for u in range(V):
        for v in adj[u]:
            
            if u == v:
                continue # self-loops
            
            edge = (min(u, v), max(u, v))
            
            if edge not in seen:
                seen.add(edge)
                adj_prime[u].append(v)
                adj_prime[v].append(u)
    
    return adj_prime
```

```python
def square_adj_list(adj):
    V = len(adj)
    adj2 = [set() for _ in range(V)]
    
    for u in range(V):
        for v in adj[u]:
            adj2[u].add(v)
            
            for w in adj[v]:
                adj2[u].add(w)
    
    return [list(neighbors) for neighbors in adj2]
```

```python
def square_matrix(A):
    V = len(A)
    A2 = [[0]*V for _ in range(V)]
    
    for i in range(V):
        for j in range(V):
            if A[i][j] == 1:
                A2[i][j] = 1  # distance 1
            
            else:
                for k in range(V):
                    if A[i][k] == 1 and A[k][j] == 1:
                        A2[i][j] = 1
                        break
    
    return A2
```

```python
"""
In a directed graph, a universal sink is a vertex that:
has no outgoing edges (points to nobody)
has incoming edges from everyone else

In matrix terms (for vertex s):
Row s = all 0s
Column s = all 1s (except diagonal)

    0 1 2 3
  ---------
0 | 0 1 1 1
1 | 0 0 1 1
2 | 0 0 0 0   ← candidate?
3 | 0 1 1 0

A[u][v] = 1 means u → v
A[u][v] = 0 means no edge

"eliminate" This vertex cannot be the sink anymore, so ignore it.

Case 1:
A[i][j] == 1
means:
i → j
So i cannot be a sink (because sinks have no outgoing edges)

Case 2:
A[i][j] == 0
means:
i does NOT point to j
So j cannot be a sink (because sinks must have incoming from everyone)
Eliminate column j → move right:
j += 1

Step-by-step:
Step 1: A[0][0] = 0 → eliminate column 0 → j = 1
Step 2: A[0][1] = 1 → eliminate row 0 → i = 1
Step 3: A[1][1] = 0 → eliminate column 1 → j = 2
Step 4: A[1][2] = 1 → eliminate row 1 → i = 2
Step 5: A[2][2] = 0 → eliminate column 2 → j = 3
Step 6: A[2][3] = 0 → eliminate column 3 → j = 4 (stop)

0 → 2
1 → 2
3 → 2

2 → (no one)

Walk right on 0, walk down on 1 — whoever survives might be the sink.
"""
def find_candidate(A):
    n = len(A)
    i, j = 0, 0
    
    while i < n and j < n:
        if A[i][j] == 1:
            i += 1
        else:
            j += 1
    
    return i if i < n else -1

def is_universal_sink(A, c):
    n = len(A)
    
    for j in range(n):
        if A[c][j] != 0:
            return False
    
    for i in range(n):
        if i != c and A[i][c] != 1:
            return False
    
    return True

def universal_sink(A):
    c = find_candidate(A)
    
    if c == -1:
        return None
    
    return c if is_universal_sink(A, c) else None
```

```python
"""
GRAPH:
1 → 2
1 → 3

VERTICES: {1,2,3}
EDGES: e1=(1→2), e2=(1→3)

INCIDENCE MATRIX B (rows=vertices, cols=edges):
        e1  e2
      --------
1 |    -1  -1
2 |     1   0
3 |     0   1

TRANSPOSE Bᵀ:
        1   2   3
      -----------
e1 |  -1   1   0
e2 |  -1   0   1

COMPUTE BBᵀ (row dot products):
(1,1): [-1,-1]·[-1,-1] = 2
(1,2): [-1,-1]·[ 1, 0] = -1
(1,3): [-1,-1]·[ 0, 1] = -1
(2,2): [ 1, 0]·[ 1, 0] = 1
(2,3): [ 1, 0]·[ 0, 1] = 0
(3,3): [ 0, 1]·[ 0, 1] = 1

inverse:
(2,1) = (1,2) = -1
(3,1) = (1,3) = -1
(3,2) = (2,3) = 0

BBᵀ = (BBᵀ)ᵀ

FINAL RESULT:
BBᵀ =
[ 2  -1  -1
 -1  1   0
 -1  0   1 ]

        1   2   3
      -----------
1 |    2  -1  -1
2 |   -1   1   0
3 |   -1   0   1

FASTER METHOD (no multiplication):
For each edge u→v:
(u,u)+=1, (v,v)+=1
(u,v)+=-1, (v,u)+=-1
Compute upper triangle → mirror it

KEY IDEA:
BBᵀ[i][i] = degree of vertex i
BBᵀ[i][j] = - (# edges between i and j)
"""
```

```python
"""
graph = {
    'A': {'adj': ['B', 'C']},
    'B': {'adj': ['A', 'D', 'E']},
    'C': {'adj': ['A', 'F']},
    'D': {'adj': ['B']},
    'E': {'adj': ['B', 'F']},
    'F': {'adj': ['C', 'E']}
}

Nodes turn GRAY when discovered (added to queue)
Nodes turn BLACK when done
Distance = number of edges from start

Think of BFS like cooking pancakes on a conveyor belt

WHITE → batter not poured yet
GRAY → pancake cooking (on the belt)
BLACK → pancake done

Queue = conveyor belt
→ first pancake in = first out

BFS discovers nodes in increasing order of distance from the source
"""
from collections import deque

def bfs(g, s):
    for u in g:
        u.color = 'WHITE'
        u.d = float("inf")
        u.pi = None

    s.color = 'GRAY'
    s.d = 0
    s.pi = None

    q = deque()
    q.append(s)

    while q:
        u = q.popleft()
        for v in u.adj:
            if v.color == 'WHITE':
                v.color = 'GRAY'
                v.d = u.d + 1
                v.pi = u
                q.append(v)
        u.color = 'BLACK'
```

```python
"""
A: [B, C]
B: [D]
C: []
D: []

Alphabetical order → A, B, C, D

(node, next_neighbor_index)

Starting with A, the stack begins as [(A, 0)] and A is discovered
we then explore its first neighbor B,
updating A to (A, 1) and pushing B → [(A, 1), (B, 0)]. 
From B we go to D, updating B to (B, 1) and pushing D → [(A, 1), (B, 1), (D, 0)].
D has no neighbors, so it finishes and is popped → [(A, 1), (B, 1)]
B also has no more neighbors, so it finishes and is popped → [(A, 1)].
We return to A and explore its next neighbor C, updating A to (A, 2) and pushing C → [(A, 2), (C, 0)].
C has no neighbors, so it finishes and is popped → [(A, 2)], and finally A finishes and is popped, leaving the stack empty.

So:
(A, 0) = A hasn’t explored any neighbors yet
(A, 1) = finished first neighbor, moving to next

(A, 0) → go to B
(A, 1) → go to C
(A, 2) → done

The stack is recursion, frozen in time.

Each element remembers:
where you are (u)
what you were about to do next (i)
"""
def dfs(graph):
    color = {u: "WHITE" for u in graph}
    time = 0

    for start in sorted(graph):
        if color[start] == "WHITE":
            stack = [(start, 0)]  # (node, next neighbor index)

            while stack:
                u, i = stack[-1]

                if color[u] == "WHITE":
                    color[u] = "GRAY"
                    time += 1
                    print(f"discover {u} at {time}")

                # still has neighbors to explore
                if i < len(graph[u]):
                    v = graph[u][i]
                    stack[-1] = (u, i + 1)

                    if color[v] == "WHITE":
                        stack.append((v, 0))

                else:
                    stack.pop()
                    color[u] = "BLACK"
                    time += 1
                    print(f"finish {u} at {time}")
```

```python
"""
Graph Type	Edge Types
Directed:	Tree, Back, Forward, Cross
Undirected:	Tree, Back only

Edge type = what state the neighbor is in when you see it
DFS doesnt care what edges exist it only cares whats still WHITE.
"""
def dfs(graph):
    color = {u: "WHITE" for u in graph}
    d = {}
    f = {}
    time = 0

    for u in sorted(graph):
        if color[u] == "WHITE":
            dfs_visit(graph, u, color, d, f, time)


def dfs_visit(graph, u, color, d, f, time):
    color[u] = "GRAY"
    time += 1
    d[u] = time

    for v in graph[u]:
        if color[v] == "WHITE":
            print(f"{u} → {v} : TREE EDGE")
            dfs_visit(graph, v, color, d, f, time)

        elif color[v] == "GRAY":
            print(f"{u} → {v} : BACK EDGE")

        elif color[v] == "BLACK":
            # Need timestamps to distinguish
            if d[u] < d[v]:
                print(f"{u} → {v} : FORWARD EDGE")
            else:
                print(f"{u} → {v} : CROSS EDGE")

    color[u] = "BLACK"
    time += 1
    f[u] = time
```

```python
def traverse_all_edges(graph):
    visited = set()
    path = []

    def dfs(u):
        visited.add(u)

        for v in graph[u]:
            if v not in visited:
                path.append((u, v))  # forward
                dfs(v)
                path.append((v, u))  # backward

    start = next(iter(graph))
    dfs(start)

    return path
```

```python
"""
graph = {
    'p': ['o', 's'],
    'o': ['r', 'v'],
    's': ['r'],
    'r': ['y'],
    'y': ['v'],
    'v': []
}

print(count_paths_dag(graph, 'p', 'v')) : 4

p → o, s
o → r, v
s → r
r → y
y → v
v → []

indegress:
p: 0
o: 1   (p → o)
s: 1   (p → s)
r: 2   (o → r, s → r)
y: 1   (r → y)
v: 2   (o → v, y → v)

Process in topo order:
p
o += 1 → 1
s += 1 → 1
o
r += 1 → 1
v += 1 → 1
s
r += 1 → 2
r
y += 2 → 2
y
v += 2 → 3

"""
from collections import defaultdict, deque

def count_paths_dag(graph, a, b):
    # graph: adjacency list {u: [v1, v2, ...]}
    
    # in-degrees number of verticies pointing to
    # d['x'] += 1 works starts at 0 automatically instead of: 
    # d = {}, d['x'] = 0, d['x'] += 1
    indegree = defaultdict(int)
    for u in graph:
        for v in graph[u]:
            indegree[v] += 1
        if u not in indegree:
            indegree[u] = indegree[u]
    
    # Step 2: topological sort (Kahn's algorithm)
    queue = deque([v for v in indegree if indegree[v] == 0])
    topo = []
    
    while queue:
        u = queue.popleft()
        topo.append(u)
        for v in graph[u]:
            indegree[v] -= 1
            if indegree[v] == 0:
                queue.append(v)
    
    paths = {v: 0 for v in indegree}
    paths[a] = 1
    
    for u in topo:
        for v in graph[u]:
            paths[v] += paths[u]
    
    return paths[b]
```

```python
"""
init:

time = 0
bcc_id = 0
stack = []

for each edge e:
    e.bcc = None

    1
   / \
  2---3
   \
    4

BCC 1: (1,2), (2,3), (3,1)
BCC 2: (2,4)

(1,2).bcc = 1
(2,3).bcc = 1
(3,1).bcc = 1

(2,4).bcc = 2

DFS + stack groups edges into "chunks"
Each chunk = one biconnected component
We just assign a unique ID to each chunk
"""
def DFS(v, parent):
    global time, bcc_id
    v.d = time
    v.low = v.d
    time += 1

    for w in neighbors(v):
        if w.d is None:
            stack.append((v, w))
            DFS(w, v)
            v.low = min(v.low, w.low)

            if w.low >= v.d:
                bcc_id += 1

                while True:
                    e = stack.pop()
                    e.bcc = bcc_id
                    if e == (v, w):
                        break

        elif w != parent and w.d < v.d:
            stack.append((v, w))
            v.low = min(v.low, w.d)
```
