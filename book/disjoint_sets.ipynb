```python
class UnionFind:
    def __init__(self, vertices):
        self.parent = {v: v for v in vertices}

    def find(self, x):
        if self.parent[x] != x:
            self.parent[x] = self.find(self.parent[x])  # Path compression
        return self.parent[x]

    def union(self, x, y):
        rootX = self.find(x)
        rootY = self.find(y)
        if rootX != rootY:
            self.parent[rootY] = rootX

    def get_components(self):
        comps = {}
        for v in self.parent:
            root = self.find(v)
            if root not in comps:
                comps[root] = []
            comps[root].append(v)
        return list(comps.values())


# Given data
V = ['a','b','c','d','e','f','g','h','i','j','k']
edges = [
    ('d','i'), ('f','k'), ('g','i'), ('b','g'), ('a','h'),
    ('i','j'), ('d','k'), ('b','j'), ('d','f'), ('g','j'), ('a','e')
]

uf = UnionFind(V)

# Process edges step by step
for step, (u, v) in enumerate(edges, 1):
    uf.union(u, v)
    components = uf.get_components()
    
    # Sort each component for readability
    components = [sorted(comp) for comp in components]
    
    print(f"Step {step} after edge ({u}, {v}):")
    print(components)
    print("-" * 40)
```

```python
"""
a = make_set('a')
b = make_set('b')
c = make_set('c')

union(a, b)
union(a, c)

a → b → c
"""
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
        self.rep = None  # pointer to the set (Set object)


class SetObject:
    def __init__(self, node):
        self.head = node
        self.tail = node
        self.size = 1

def make_set(x):
    node = Node(x)
    s = SetObject(node)
    node.rep = s
    return node

def find_set(x):
    return x.rep


# UNION(x, y) using weighted-union
def union(x, y):
    set_x = find_set(x)
    set_y = find_set(y)

    if set_x == set_y:
        return set_x

    # Ensure set_x is the larger set
    if set_x.size < set_y.size:
        set_x, set_y = set_y, set_x

    set_x.tail.next = set_y.head

    current = set_y.head
    while current:
        current.rep = set_x
        current = current.next

    set_x.tail = set_y.tail
    set_x.size += set_y.size

    return set_x
```

```python
"""
Disjoint-set forest with union by rank and path compression
"""
def make_set(x):
    x.p = x
    x.rank = 0

def find_set(x):
    if x != x.p:
        x.p = find_set(x.p)  # path compression
    return x.p

def union(x, y):
    link(find_set(x), find_set(y))

def link(x, y):
    if x.rank > y.rank:
        y.p = x
    elif x.rank < y.rank:
        x.p = y
    else:
        y.p = x
        x.rank += 1
```

```python
"""
parent = {1:1, 2:1, 3:1, 4:3}
print(find_set(parent, 4))  # → 1
"""
def find_set(parent, x):
    root = x
    
    while root != parent[root]:
        root = parent[root]

    # Second pass: path compression
    while x != root:
        next_node = parent[x]
        parent[x] = root
        x = next_node

    return root

def print_set(x):
    root = find_set(x)
    curr = root.head
    while curr is not None:
        print(curr)
        curr = curr.next # next is new here for print_set
```

```python
"""
in a standard disjoint-set forest, each node only keeps track of its parent (and sometimes its rank for union by rank).
There’s no need to store the whole tree structure or children, because:

The parent pointer alone defines the tree implicitly.
find_set climbs parents until it reaches a root.
union only changes the parent of one root to point to another root.
Path compression just rewires parent pointers along the path to the root.

So after flattening, even though the tree looks like a “star,” no extra pointers are stored
the structure exists entirely through parent links.

The only optional extra attribute you might add is rank (for union by rank),
or next if you want to print all members of a set efficiently.
Otherwise, parent alone is enough for union-find operations.
"""
```

```python
"""
Levels stay small because the Ackermann function grows extremely fast, while ranks grow very slowly.

Ranks grow slowly (≤ log n)
Ackermann explodes instantly
→ k must stay tiny

Ackermann = jumps like: 3 → 4 → 16 → 65536 → galaxy-sized

ACKERMANN FUNCTION – GROWTH LEVELS

A₀(n) = n + 1                      (linear)
A₁(n) ≈ n + 2                      (linear)
A₂(n) ≈ 2n                         (linear, faster)
A₃(n) ≈ 2ⁿ                         (exponential)
A₄(n) ≈ power tower of 2s height n (extremely fast)

---

INTUITION (OPERATIONS STACKING)

A₀ → addition
A₁ → repeated addition
A₂ → multiplication
A₃ → exponentiation
A₄ → repeated exponentiation (towers)

---

GROWTH TABLE

## n    A₀(n)   A₁(n)   A₂(n)   A₃(n)=2ⁿ     A₄(n)

1      2       3       2        2           4
2      3       4       4        4           16
3      4       5       6        8           65536
4      5       6       8        16          2^65536 (huge)

---

INVERSE ACKERMANN α(n)

α(n) = largest k such that A_k(1) ≤ n

VALUES IN PRACTICE:

n ≤ 4                  → α(n) = 2
n ≤ 16                 → α(n) = 3
n ≤ 2^65536            → α(n) = 4
n ≤ anything realistic → α(n) ≤ 4

---

KEY FACT

Ackermann grows so fast that α(n) grows extremely slowly.

In all practical cases:
α(n) ≤ 4

---

APPLICATION (UNION-FIND)

level(x) = max k such that A_k(rank(x)) ≤ rank(parent(x))

Because A_k grows extremely fast:
→ k can only be very small

Therefore:
0 ≤ level(x) < α(n) ≤ 4 (in practice)

---

INTUITION

k = 0 → small increase
k = 1 → small increase
k = 2 → moderate
k = 3 → big jump
k = 4 → explodes beyond any realistic number

Path compression forces nodes upward so aggressively that each node can only incur cost α(n) times → total time O(n α(n))

depth(x) = number of edges to root   ↓ decreases
level(x) = Ackermann-based measure   ↑ increases

Compression reduces depth,
but increases level.

Path compression flattens the tree (depth ↓),
but increases level because parent ranks become much larger.

"""
```

```python
"""
Pseudodistance = lazy propagation of depth information.

Only the grafted root changes its value
All descendants are handled automatically via sum along parent pointers
Makes FIND-DEPTH + GRAFT very efficient, near-linear overall
"""
```

```python
"""
Performs path compression, updating pseudodistances
Disjoint-set (S_i):

B (root, d=0)
|
A (p=B, d=1)
|
C (p=A, d=2)

FIND_DEPTH(C) = C.d + A.d + B.d = 2 + 1 + 0 = 3
FIND_DEPTH(A) = A.d + B.d = 1 + 0 = 1
FIND_DEPTH(B) = 0
"""
def FIND_DEPTH(v):
    if v.p == v:
        # v is root
        return 0
    else:
        parent_depth = FIND_DEPTH(v.p)
        v.d += v.p.d
        v.p = v.p.p
        return v.d

def GRAFT(r, v):
    r.d = 1 + FIND_DEPTH(v) - FIND_DEPTH(r)
    r.p = v
```

```python
"""
we already updated r.d before linking → depths remain correct
Only root of grafted tree needs pseudodistance update

Aspect	Previous discussion	This union/link version
Parent pointer update	r.p = v	r.p updated via LINK(x, y)
Depth update	r.d = 1 + FIND_DEPTH(v) - FIND_DEPTH(r)	Same: r.d = 1 + FIND_DEPTH(v) - FIND_DEPTH(r)
Efficiency	Works but may not use rank	Uses union-by-rank → better amortized complexity
Framing	Operational, direct	Union-Find canonical (FIND/UNION/LINK)
"""
def GRAFT(r, v):
    root_r = FIND_SET(r)
    root_v = FIND_SET(v)

    if root_r == root_v:
        return

    r.d = 1 + FIND_DEPTH(v) - FIND_DEPTH(r)

    LINK(root_r, root_v)

def LINK(x, y):
    if x.rank > y.rank:
        y.p = x
    else:
        x.p = y
        if x.rank == y.rank:
            y.rank += 1
```

```python
"""
Tree T:
        A
       / \
      B   C
     / \
    D   E

Queries P:
{D, E}
{D, C}

DFS Traversal Order (one possible order):
A → B → D → E → C

1. Visit D:
   - D has no children
   - D.color = BLACK
   - ancestor[D] = D

2. Visit E:
   - E has no children
   - E.color = BLACK

   Check query {D, E}:
   - D is BLACK
   → LCA(D, E) = ancestor[find_set(D)] = B

3. Back to B:
   - union(B, D), union(B, E)
   - ancestor[B] = B
   - B.color = BLACK

4. Visit C:
   - C has no children
   - C.color = BLACK

   Check query {D, C}:
   - D is BLACK
   → LCA(D, C) = ancestor[find_set(D)] = A

LCA(D, E) = B
LCA(D, C) = A

- Each pair {u, v} is checked twice (once at u, once at v)
- The LCA is computed only when the second node becomes BLACK
- ancestor[find_set(x)] always gives the correct LCA at that moment

Each level of depth corresponds to one “active” set during DFS.
"""
class Node:
    def __init__(self, name):
        self.name = name
        self.parent = self      # DSU parent
        self.rank = 0
        self.ancestor = None
        self.color = "WHITE"
        self.children = []      # tree edges

def find_set(x):
    if x.parent != x:
        x.parent = find_set(x.parent)
    return x.parent


def union(x, y):
    x_root = find_set(x)
    y_root = find_set(y)

    if x_root == y_root:
        return

    if x_root.rank < y_root.rank:
        x_root.parent = y_root
    else:
        y_root.parent = x_root
        if x_root.rank == y_root.rank:
            x_root.rank += 1


def lca(u, queries, answers):
    make_set(u)
    u.parent = u
    u.rank = 0
    find_set(u).ancestor = u

    for v in u.children:
        lca(v, queries, answers)
        union(u, v)
        find_set(u).ancestor = u

    u.color = "BLACK"

    for v in queries[u]:
        if v.color == "BLACK":
            lca_node = find_set(v).ancestor
            answers[(u.name, v.name)] = lca_node.name
```
