```python
class UnionFind:
    def __init__(self, vertices):
        self.parent = {v: v for v in vertices}

    def find(self, x):
        if self.parent[x] != x:
            self.parent[x] = self.find(self.parent[x])  # Path compression
        return self.parent[x]

    def union(self, x, y):
        rootX = self.find(x)
        rootY = self.find(y)
        if rootX != rootY:
            self.parent[rootY] = rootX

    def get_components(self):
        comps = {}
        for v in self.parent:
            root = self.find(v)
            if root not in comps:
                comps[root] = []
            comps[root].append(v)
        return list(comps.values())


# Given data
V = ['a','b','c','d','e','f','g','h','i','j','k']
edges = [
    ('d','i'), ('f','k'), ('g','i'), ('b','g'), ('a','h'),
    ('i','j'), ('d','k'), ('b','j'), ('d','f'), ('g','j'), ('a','e')
]

uf = UnionFind(V)

# Process edges step by step
for step, (u, v) in enumerate(edges, 1):
    uf.union(u, v)
    components = uf.get_components()
    
    # Sort each component for readability
    components = [sorted(comp) for comp in components]
    
    print(f"Step {step} after edge ({u}, {v}):")
    print(components)
    print("-" * 40)
```

```python
"""
a = make_set('a')
b = make_set('b')
c = make_set('c')

union(a, b)
union(a, c)

a → b → c
"""
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
        self.rep = None  # pointer to the set (Set object)


class SetObject:
    def __init__(self, node):
        self.head = node
        self.tail = node
        self.size = 1

def make_set(x):
    node = Node(x)
    s = SetObject(node)
    node.rep = s
    return node

def find_set(x):
    return x.rep


# UNION(x, y) using weighted-union
def union(x, y):
    set_x = find_set(x)
    set_y = find_set(y)

    if set_x == set_y:
        return set_x

    # Ensure set_x is the larger set
    if set_x.size < set_y.size:
        set_x, set_y = set_y, set_x

    set_x.tail.next = set_y.head

    current = set_y.head
    while current:
        current.rep = set_x
        current = current.next

    set_x.tail = set_y.tail
    set_x.size += set_y.size

    return set_x
```

```python
"""
Disjoint-set forest with union by rank and path compression
"""
def make_set(x):
    x.p = x
    x.rank = 0

def find_set(x):
    if x != x.p:
        x.p = find_set(x.p)  # path compression
    return x.p

def union(x, y):
    link(find_set(x), find_set(y))

def link(x, y):
    if x.rank > y.rank:
        y.p = x
    elif x.rank < y.rank:
        x.p = y
    else:
        y.p = x
        x.rank += 1
```
