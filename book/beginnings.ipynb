```python
"""
j walks left;
every big number j sees gets pushed right;
when j stops, the key drops into the hole.
"""
input_list = [31, 41, 59, 26, 41, 58]

def insertion_sort(a):
  for i in range(1, len(a)):
    key = a[i]
    j = i - 1
    while j >= 0 and a[j] > key:
      a[j + 1] = a[j]
      j -= 1
    a[j + 1] = key
  return a
    
print(insertion_sort(input_list))
# sorted(input_list)
```

```python
"""
first need to split array into L and R from A[p:r]
A[p:q] to A[q+1:r]

This says if we have values in a middle of list (a slice of an array)
nL = q - p + 1 (length of A[p:q])
nR = r - q (length of A[q+1:r])

for this example we can just use the whole list making:
p = 0, q = Math.floor(len(A) / 2), r = len(A)


"""
input_list = [3, 41, 52, 26, 38, 57, 9, 49]

def merge(a, p, q, r):
    L = a[p:q]
    R = a[q:r]
    i = j = 0
    k = p
    while i < len(L) and j < len(R):
        if L[i] <= R[j]:
            a[k] = L[i]
            i += 1
        else:
            a[k] = R[j]
            j += 1
        k += 1
    while i < len(L):
        a[k] = L[i]
        i += 1
        k += 1
    while j < len(R):
        a[k] = R[j]
        j += 1
        k += 1

def merge_sort(a, p, r):
    if r - p <= 1:
      return
    q = (p + r) // 2
    merge_sort(a, p, q)
    merge_sort(a, q, r)
    merge(a, p, q, r)

merge_sort(input_list, 0, len(input_list))

print(input_list)
```

```python
def merge_sort_alt(A, a = 0, b = None):
    if b in None: b = len(A)
    if 1 < b - a:
        c = (a + b + 1) // 2
        merge_sort_alt(A, a, c)
        merge_sort_alt(A, c, b)
        L, R, = A[a:c], A[c:b]
        merge_alt(L, R, A, len(L), len(R), a, b)

""" sorted L[:i] and R[:j] into A[a:b] """
def merge_alt(L, R, A, i, j, a, b):
    if a < b:
        if (j <= 0) or (i > 0 and L[i-1] > R[j-1]):
            A[b-1] = L[i-1]
            i -= 1
        else:
            A[b-1] = R[j-1]
            j -= 1
        merge(L, R, A, i, j, a, b-1)
```

```python
input_list = [3, 41, 52, 26, 38, 57, 9, 49]

def bubble_sort(a):
    for i in range(len(a)):
        for j in range(1, len(a)-i):
            if a[j] < a[j-1]:
                a[j], a[j-1] = a[j-1], a[j]
                
bubble_sort(input_list)
print(input_list)
```

```python
input_list = [3, 41, 52, 26, 38, 57, 9, 49]

def selection_sort(a):
    for i in range(len(a)):
        min_i, min_v = i, a[i]
        for j in range(i+1, len(a)):
            if a[j] < min_v:
                min_i, min_v = j, a[j]
        a[i], a[min_i] = a[min_i], a[i]

selection_sort(input_list)
print(input_list)
```

```python
def selection_sort_r(a, i = None):
    if i is None: i = len(a) - 1
    if i > 0:
        j = prefix_max(a, i)
        a[i], a[j] = a[j], a[i]
        selection_sort_r(a, i - 1)
    return a

"""
helper to return maximum a[:i + 1]
"""
def prefix_max(a, i):
    if i > 0:
        j = prefix_max(a, i- 1)
        if a[i] < a[j]:
            return j
    return i

print(selection_sort_r([8, 3, 5, 2, 9]))
```

```python
input_list = [1, 3, 4, 2, 7, 5, 6, 9, 8]

def count_long_subarray(a):
    current = 1
    length = 1
    count = 1
    for i in range(1, len(a)):
        if a[i-1] < a[i]:
            current += 1
        else:
            current = 1

        if current == length:
            count += 1
        elif current > length:
            length = current
            count = 1
    return count

print(count_long_subarray(input_list))
```

```python
input_list = [1, 3, 4, 2, 7, 5, 6, 9, 8]

def reverse(D, i, k):
    if k < 2:
        return
    x2 = D.pop(i+k-1) # delete_at()
    x1 = D.pop(i) # delete_at()
    D.insert(i, x2) # insert_at()
    D.insert(i+k-1,x1) # insert_at()
    reverse(D, i+1, k-2)
    return D

print(reverse(input_list, 0, 5))
```

```python
"""
Move k items starting at i to be in the front
of item at index j
"""
input_list = [1, 3, 4, 2, 7, 5, 6, 9, 8]

def move(D, i, k, j):
    if k < 1:
        return
    x = D.pop(i) # delete_at()
    if j > i:
        j -= 1
    D.insert(j, x) # insert_at()
    if i > j:
        i += 1
    move(D, i, k-1, j)
    return D

print(move(input_list, 3, 5, 0))
```

```python
class Doubly_LL_Node:
    def __init__(self):
        return

class Doubly_LL:
    def __init__(self, item):
        self.item = item
        self.next = None
        self.prev = None

    def first(self):
        return self.head

    def last(self):
        return self.last

    def next(self, x):
        return x.next

    def prev(self, x):
        return x.prev

    def is_empty(self):
        return self.head is None
    
    def insert_first(self, x):
        new_node = Doubly_LL_Node(x)
        if self.head is None:
            self.head = new_node
            self.tail = new_node
        else:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node

    def insert_last(self, x):
        new_node = Doubly_LL_Node(x)
        if self.tail is None:
            self.head = new_node
            self.tail = new_node
        else:
            new_node.prev = self.tail
            self.tail.next = new_node
            self.tail = new_node

    def delete_first(self):
        assert self.head
        x = self.head.item
        self.head = self.head.next
        if self.head is None: self.tail = None
        else: self.head.prev = None
        return x

    def delete_last(self):
        assert self.tail
        x = self.tail.item
        if self.tail is None: self.head = None
        else: self.tail.next = None
        return x

    def remove(self, x1, x2):
        L2 = Doubly_LL()
        L2.head = x1
        L2.tail = x2
        if x1 == self.head: self.head = x2.next
        else: x1.prev.next = x2.next
        if x2 == self.tail: self.tail = x1.prev
        else: x2.next.prev = x1.prev
        x1.prev = None
        x2.next = None
        return L2

    def splice(self, x, L2):
        xn = x.next
        x1 = L2.head
        x2 = L2.tail
        L2.head = None
        L2.tail = None
        x1.prev = x
        x.next = x1
        x2.next = xn
        if xn: xn.prev = x2
        else: self.tail = x2
```
